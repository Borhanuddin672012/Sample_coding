```
import pandas as pd
```

```
df = pd.read_csv('carprices.csv')
df
```

```
dumies = pd.get_dummies(df['Car Model'])
dumies
```

```
#marged = pd.concat([df, dumies],axis='colums')
#marged = pd.concat([df, dumies],axis=1)
marged = pd.concat([df, dumies])

marged
```

```
#marged = pd.concat([df, dumies],axis='colums')
marged = pd.concat([df, dumies],axis=1)
#marged = pd.concat([df, dumies])

marged
```

```
final = marged.drop(['Car Model'], axis=1)
```

```
final
```

```
y = df['Sell Price']
```

```
y
```

```
x = final.drop(['Sell Price'], axis=1)
```

```
x
```

```
from sklearn.linear_model import LinearRegression
```

```
model = LinearRegression()
```

```
model.fit(x,y)
```

```
model.score(x,y)
```

```
model.predict([[45000,8,1,0,0,0]])
```
